```python
#pour faire des opérations, on supprime l'entête avec le parametre skip_header=1
world_alcohol = np.genfromtxt("/home/francis/Téléchargements/world_alcohol.csv", delimiter=',', dtype = "U75", skip_header=1)
world_alcohol
```

```python

# afficher toutes les lignes contenant 'Algeria' dans la 3ème colonne dans le df

country_is_vietnam = world_alcohol[:,2] == 'Viet Nam'
```

```python
ar_vietnam = world_alcohol[country_is_vietnam]
ar_vietnam
```

```python
import pandas as pd
df_vietnam = pd.DataFrame(ar_vietnam)
df_vietnam
```

```python
headers = ["years", "continent","country", "type_of_alcohol", "liters"] # définition de l'entête
df_vietnam.columns = headers # ajout de l'entête
df_vietnam
```

```python
df_vietnam.dropna(subset=['liters'], axis=0, inplace=True)
df_vietnam
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

name = df_vietnam['years']
liters = df_vietnam['liters']
fig =plt.figure(figsize = (10, 7))

font1 ={'family':'serif', 'color':'darkred', 'size':25}
font2 = {'family':'serif', 'size':20}
plt.bar(name[0:10], liters[0:10])
plt.title("Consommation moyenne d'alcool au Vietnam par habitant", fontdict=font1)
plt.xlabel('Année')
plt.ylabel('litres')
plt.show
```
